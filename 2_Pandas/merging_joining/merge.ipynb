```python
import pandas as pd
```

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
pd.concat([week1,week2])

#will join the values from week2 below week 1
```

```python
pd.concat([week1,week2], keys = ["Week1", "Week2"])
#will join the values from week2 below week1 but the keys parameter will give an index value to distinguish between week1 n week2 values
```

```python
week1.merge(week2, how = "inner", on = "Customer ID")
#will merge only the common values value from week 1 and week 2  
```

```python
week1.merge(week2, how = "inner", on = "Customer ID", suffixes= [" - Week1", "- Week2"])
```

```python
week1[week1["Customer ID"] == 155]
#will give the rows relatd to customer id 155
```

```python
week1.merge(week2, how = "inner", on = ["Customer ID", "Food ID"])
#will give only those common values which occur in customer id and food id in both weeks
```

```python
week1.merge(week2, how = "outer", on = "Customer ID")
#will give either single occuring or the values occuring both times
```

```python
week1.merge(foods, how = "left", on = "Food ID")
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID", sort = True)
#.drop("ID", axis = "columns")
```
